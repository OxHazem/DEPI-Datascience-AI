```python
#begining of Viusalization 
#1 overall score distribution
sns.histplot(data['GPA'],kde=True,bins=25 ,label='GPA',alpha=0.5 , linewidth=1)
```

```python
#2 What is the average GPA per parental education level
grouppeddata=data.groupby('ParentalEducation')['GPA'].mean()
plt.figure(figsize=(10,6))

plt.pie(grouppeddata,labels=grouppeddata.index,autopct='%1.1f%%',startangle=140)
plt.legend(grouppeddata.index,loc='upper right')
plt.title("Average GPA by Parental Education Level", fontsize=16)
plt.show()
```

```python
#3 study time vs GPA
plt.figure(figsize=(10,6))
sns.scatterplot(x="StudyTimeWeekly", y="GPA", data=data)
plt.title("Study Time vs GPA", fontsize=16)
plt.show()
```

```python
#4 Absences vs GPA
plt.figure(figsize=(10,6))
sns.lineplot(x="Absences", y="GPA", data=data)
plt.title("Absences vs GPA", fontsize=16)
plt.show()
```

```python
#5 Tutoring vs GPA
grouppeddata=data.groupby('Tutoring')['GPA'].mean()
labels = ["No", "Yes"]
plt.figure(figsize=(10,6))
plt.pie(grouppeddata,labels=labels,autopct='%1.1f%%',startangle=140)
plt.legend(grouppeddata.index,loc='upper right',fontsize=9,title='Tutoring',title_fontsize='9',labels=labels)
plt.show()

```
